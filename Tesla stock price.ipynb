```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv(r'D:\Files\TSLA.csv')
```

```python
df.info()
```

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
df['Year']=df['Date'].dt.year
```

```python
df['Month']=df['Date'].dt.month
```

```python
round(df.describe())
```

```python
plt.style.use("ggplot")
plt.figure(figsize=(16,6))
sns.lineplot(x='Date',y='Open',data=df,color='Red',label='Open')
sns.lineplot(x='Date',y='Close',data=df,color='Blue',label='Close')
plt.title('The relation between Date of Open & Close Value')
```

```python
plt.style.use("ggplot")
plt.figure(figsize=(16,6))
sns.lineplot(x='Date',y='Low',data=df,color='Red',label='Low')
sns.lineplot(x='Date',y='High',data=df,color='Blue',label='High')
plt.title('The relation between Date of Low & High Value')
```

```python
plt.figure(figsize=(16,6))
sns.scatterplot(x='Date',y='Volume',data=df)
plt.title('The relation for Stock Price')
```
